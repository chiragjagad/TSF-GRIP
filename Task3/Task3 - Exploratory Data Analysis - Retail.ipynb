## Data Visualization

### Histogram

```python
dataset.hist(figsize=[10,10])
plt.show()
```

### Heatmap

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.heatmap(dataset.corr(method='pearson'), annot=True,ax=ax, vmin=-1, vmax=1, center=0);
```

```python
# Making a dataset of categorical values
dataset_cat=dataset[['Ship Mode', 'Segment', 'Country', 'City', 'State',
       'Region', 'Category', 'Sub-Category']]
```

### Count Plots

```python
for col in dataset_cat.columns:
    if(col=='State' or col=='City'):
        plt.figure(figsize=(15,8))
    sns.barplot(dataset_cat[col].value_counts()[:50].index,dataset_cat[col].value_counts()[:50]).set_title(col)
    plt.xticks(rotation=90)
    plt.show()
```

### Pie Charts

```python
for col in dataset_cat.columns:
    fig = plt.figure(figsize =(7, 7)) 
    plt.pie(dataset[col].value_counts(), labels = dataset[col].value_counts().index, autopct='%1.1f%%')
    plt.title(col)
    plt.show()
```

### Profits and Sales analysis

```python
for col in dataset_cat.columns:
    if(col=='State' or col =='City'):
            dataset.groupby(dataset[col])['Profit','Sales'].sum().plot(kind = 'bar', figsize=(12,8))
            continue
    dataset.groupby(dataset[col])['Profit','Sales'].sum().plot(kind = 'bar')
    plt.show()
```

### Statastics based on Discount

```python
for col in dataset_cat.columns:
    mean_dataset = dataset.groupby(dataset[col])['Discount'].agg(['mean','sum','max','min','count'])
    mean_dataset = mean_dataset.reset_index()
    print(col)
    print(mean_dataset)
    print()
```

### Pairplot

```python
figsize=(15,10)
sns.pairplot(dataset,hue='Sub-Category')
```

## Observations

### Shipping mode wise

1. The standard class has the highest profits and sales, but the ratio of two is far too less. Whereas, the ratio in the first class and the second class is relatively higher. So, the company should focus on attracting customers for the first class and the second class delivery

### State wise

1. The company is doing pretty well in the prominent states of the United States. But it is lacking a lot in various states like North Dakota, West Virginia, Wyoming, Maine, South Dakota, District of Columbia, Vermont, Idaho, Montana, etc. The deals here are negligible. The company should try to understand the customers need in these states and increase the deals

### Region wise

1. We found that sales are minimum in the southern region of the United States, so the company should focus on the southern region to grow sales
2. Another observation we could see is that sales are notable in the central region, but profits are minimum. A possible reason could be the higher discounts in the central region. Thus, the company should concentrate on building profits in the central region

### Category wise

1. The company has a good number of deals with Office Supplies. But it isn't able to generate any profits. Generating profits through these products could be the key!

2. Furniture too has a good number of deals, but profit is negligible. So, the company should focus on this category too

### Subcategory wise

1. We can see that copiers and machines have the lowest number of deals. But sales and profits are much higher. So the company should concentrate on increasing their deals

2. Another thing we could see is that the company is selling tables at a loss which the company should not. Selling tables with lesser discounts and focusing on its deals could generate profits
